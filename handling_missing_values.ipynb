## handling missing values
- `df,insull()` - to detect missing vals(`isna()` is the same thing)
- `df,fillna()` - to fill missing vals(imputation)
- `df.dropna()` - to remove missing vals
- `df.replce()` - to replace specific vals(imputation)

```python
import numpy as np
import pandas as pd
```

```python
employee_list = [['James', 36, 75, 5428000],
               ['Villers', 38, 74, 3428000],
               ['VKole', 31, 70, 8428000],
               ['Smith', 34, 80, 4428000],
               ['Gayle', 40, 100, 4528000],
               ['Adam', 40, np.nan, 4528000],
               ['Rooter', 33, 72, 7028000],
               ['Peterson', 42, 85, 2528000],
               ['lynda', 42, 85, np.nan],
               [np.nan, 42, 85, np.nan],
               ['Jenny', np.nan, 100, 25632],
               ['Kenn', np.nan, 110, 25632],
                ['Aly', np.nan, 90, 25582],
               ['John', 41, 85, 1528000]]



df = pd.DataFrame(employee_list, columns=['name', 'age', 'weight', 'salary'])
```

```python
rows, cols = df.shape

print(f'Rows: {rows}')
print(f'Cols: {cols}')
```

```python
df.head()
```

```python
df.isnull().sum()
```

```python
dropped_df = df.dropna()
# when dropping na, you remove the entire row

print(dropped_df)
```

```python
filled_df = df.fillna('None')



print(filled_df)
```

```python
filled_df.isnull().sum()
```

```python
sp_col = df[['weight', 'age']].fillna('Pending')

sp_col
```

```python
fd2 = df.fillna(value={'weight': 'pending', 'age': 'unknown', 'Salary': '0.0', 'Name': 'Verification Pending'})

fd2
```

```python
technologies = {
    'Courses':["Spark",np.nan,"Hadoop","Python","pandas",np.nan,"Java"],
    'Fee' :[20000,25000, np.nan,22000,24000,np.nan,22000],
    'Duration':[np.nan,'40days','35days', np.nan,'60days','50days','55days'],
    'Discount':[1000,np.nan,1500,np.nan,2500,2100,np.nan]
              }


tech = pd.DataFrame(technologies)
```

```python
rows, cols = tech.shape

print(f'Rows: {rows}')
print(f'Cols: {cols}')
```

```python
tech.isnull().sum()
```

```python
tech2 = tech.replace(np.nan, '')

tech2
```

```python
tech3 = tech.replace({'Courses': np.nan, 'Fee': np.nan}, 'NAMI')

tech3
```

```python
tech.fillna('')
```
