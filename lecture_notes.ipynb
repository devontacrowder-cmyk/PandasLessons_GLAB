## pandas datastructures
- series - a single demisional datastructure. similar to a list
    - homogeneous (of the same dt)
- dataframe - multidimensional table
    - heterogenous

```python
ser1 = pd.Series([10, 20, 30, 40], index=['A', 'B', 'C', 'D'])

print(ser1)
```

```python
d ={
    "b": 1,
    "a": 0,
    "c": 2,
    "d": 100
}

ser2 = pd.Series(d)
```

```python
d1 = {
    "a": 0.0,
    "b": 1.0,
    "c": 2.0
}

ser3 = pd.Series(d1)

print(ser3)
```

```python
calories = {
    'day1': 700,
    'day2': 2500,
    'day3': 1200
}

ser4 = pd.Series(calories)

print(ser4)
```

```python
ser5 = pd.Series(5.0, index=['a', 'b', 'c'])

print(ser5)
```

## data frame:
- indexes (each number)
- columns (each variables)

```python
fruitData = {
    'apple_sales': [3,2,0,2,9,4,7,2],
    'orange_sales': [0,3,2,7,7,8,3,0]
}

# idx = [2000, 2001, 2002, 2003, 2005, 2006, 2007, 2008]
idx = pd.RangeIndex(2000, 2008)

purchases = pd.DataFrame(fruitData, index=[idx])

print(purchases)
```

## frequency analysis - value counts

```python
data = {'Category': ['Electronics', 'Clothing', 'Electronics', 'Books', 'Books', 'Clothing']}

cat_df = pd.DataFrame(data)

cat_df.value_counts()
```

```python
sales_data={"Devices":['Laptop','iPhone','LED','LCD','Smart-Phone','Washing-Machine'],
           'Brand':['Lenovo','Apple','Samsung','Samsung','Samsung','Whirpool'],
           'Sales':[1000,2000,4000,2000,1000,4000],
           'Profit':[500,1000,1000,1500,1000,1500],
           'Pices left':[5000,4000,4000,5000,5000,1000]}


sales_df = pd.DataFrame(sales_data)

freq = sales_df['Brand'].value_counts()

freq_prod = freq[freq > 1].index[0]

print(freq)
print(f'this item appears more than once {freq_prod}')
```

```python
data = {'score': [85, 92, 88, 75, None, 90, None, 85]}

scoreDf = pd.DataFrame(data)
```

```python
scoreCounts = scoreDf.value_counts(dropna=False)
print('score', scoreCounts)
```
